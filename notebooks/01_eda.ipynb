```python
import json
import pandas as pd

with open("../data/raw/technology_news.json","r",encoding="utf-8") as f:
    articles = json.load(f)

df = pd.DataFrame(articles)
df.head()
```

```python
df.shape
df.columns
df.isnull().sum()
```

```python
df["content_length"] = df["content"].fillna("").apply(len)
df["content_length"].describe()
```

```python
import matplotlib.pyplot as plt

df["content_length"].hist()

plt.title("Article Length Distribution")
plt.xlabel("Length")
plt.ylabel("Count")
plt.show()
```
